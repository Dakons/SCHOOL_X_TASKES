```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define the dictionary we want to use
aruco_dict = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_6X6_250)

# Generate a marker
marker_id = 50
marker_size = 200  # Size in pixels
marker_image = cv2.aruco.generateImageMarker(aruco_dict, marker_id, marker_size)

cv2.imwrite('marker_42.png', marker_image)
plt.imshow(marker_image, cmap='gray', interpolation='nearest')
plt.axis('off')  # Hide axes
plt.title(f'ArUco Marker {marker_id}')
plt.show()
```

```python
import cv2
import numpy as np

# Инициализация веб-камеры
cap = cv2.VideoCapture(0)

# Определение словаря ArUco и параметров детектора
aruco_dict = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_6X6_250)
parameters = cv2.aruco.DetectorParameters()

while True:
    # Захват кадра с веб-камеры
    ret, frame = cap.read()
    if not ret:
        print("Не удалось захватить кадр")
        break

    # Преобразование кадра в оттенки серого для улучшения детекции
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Детектирование маркеров ArUco
    corners, ids, rejected = cv2.aruco.detectMarkers(gray, aruco_dict, parameters=parameters)

    # Если маркеры найдены, рисуем их на кадре
    if ids is not None:
        # Рисуем обнаруженные маркеры и их ID
        frame = cv2.aruco.drawDetectedMarkers(frame, corners, ids)
        
        # Отображаем ID и позицию маркеров
        for i in range(len(ids)):
            font = cv2.FONT_HERSHEY_SIMPLEX
            # Преобразуем угол в целочисленный формат и показываем текст на кадре
            corner = tuple(corners[i][0][0].astype(int))
            cv2.putText(frame, f"ID: {ids[i][0]}", corner, font, 0.5, (0, 255, 0), 2, cv2.LINE_AA)
            print(f"Marker ID: {ids[i][0]} detected at position {corner}")

    # Отображаем результат
    cv2.imshow("ArUco Marker Detection", frame)

    # Прерывание цикла при нажатии клавиши 'Esc'
    if cv2.waitKey(1) & 0xFF == 27:
        break

# Освобождаем камеру и закрываем окна OpenCV
cap.release()
cv2.destroyAllWindows()
```
