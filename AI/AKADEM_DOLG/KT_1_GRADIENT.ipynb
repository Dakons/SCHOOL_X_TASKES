```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

$$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Делаем массив данных
x_1 = np.linspace(start=-3, stop=3, num=500)
```

```python
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel("X",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

plt.plot(x_1,f(x_1))
plt.show()
```

Производная функции

```python
def df(x):
    return 2*x + 1
```

```python
# Построение графиков функции и её производной рядом
plt.figure(figsize=[15,5])


plt.subplot(1,2,1)

plt.xlim([-3,3])
plt.ylim(0,8)
plt.title('Функция', fontsize=17)
plt.xlabel("X",fontsize = 16)
plt.ylabel("f(x)",fontsize = 16)

plt.plot(x_1,f(x_1),color='red',linewidth=3)

# График производной функции

plt.subplot(1,2,2)
plt.title('Производная функции', fontsize=17)
plt.xlabel("X",fontsize = 16)
plt.ylabel("df(x)",fontsize = 16)
plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,6)

plt.plot(x_1,df(x_1),color='green',linewidth=3)
plt.show()
```

```python
# Градиентный спуск
new_x = 3  # Начальная точка
previous_x = 0  # Предыдущая точка
step_multiplier = 0.1  # Коэффициент шага
precision = 0.00001  # Точность, при достижении которой выполнение остановится

x_list = [new_x]  # Список значений x на каждом шаге
slope_list = [df(new_x)]  # Список значений производной (градиента) на каждом шаге

for i in range(500):  # Максимальное количество итераций - 500
    previous_x = new_x  # Обновляем предыдущее значение x
    gradient = df(previous_x)  # Вычисляем градиент
    new_x = previous_x - step_multiplier * gradient  # Обновляем значение x
#     print("previous_x = ", previous_x, ">>new_x = ",new_x)

    step_size = abs(new_x - previous_x)  # Вычисляем величину шага
#     print(step_size)

    x_list.append(new_x)  # Добавляем новое значение x в список
    slope_list.append(df(new_x))  # Добавляем значение производной в список
    
    if step_size < precision:  # Проверяем, достигнута ли требуемая точность
        print("Количество выполненных итераций:", i)
        break

print("Локальный минимум в заданном диапазоне находится в точке:", new_x)
print("Производная или df(x) в этой точке равна:", df(new_x))
print("Значение функции f(x) или функции стоимости в этой точке:", f(new_x))
```

```python
# Наложение расчетов градиентного спуска на графики

plt.figure(figsize=[20,5])

# Подграфик 1: Функция стоимости
plt.subplot(1,3,1)
plt.xlim([-3,3])
plt.ylim(0,8)
plt.title('Функция', fontsize=17)
plt.xlabel("X", fontsize=16)
plt.ylabel("f(x)", fontsize=16)

# График функции стоимости
plt.plot(x_1, f(x_1), color='red', linewidth=3, alpha=0.6)
values = np.array(x_list)  # Преобразование списка значений x в массив
plt.scatter(x_list, f(values), color='black', s=100, alpha=0.6)  # Точки на графике

# Подграфик 2: Производная функции
plt.subplot(1,3,2)
plt.title('Производная функции', fontsize=17)
plt.xlabel("X", fontsize=16)
plt.ylabel("df(x)", fontsize=16)
plt.grid()
plt.xlim(-2,3)
plt.ylim(-3,6)

# График производной функции
plt.plot(x_1, df(x_1), color='green', linewidth=3, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)  # Точки на графике


plt.show()
```
